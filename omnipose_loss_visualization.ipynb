```python
import re

import matplotlib.pyplot as plt
from pathlib import Path
```

```python
training_log_path = Path('~/omnipose_train/3D_224_train.log')
with open(training_log_path, 'r') as f:
    training_log = f.read()

epoch_pattern = re.compile(r"Train epoch: (\d+) \| Time: (\d+\.\d+)min.*<Epoch Loss>: (\d+\.\d+)")
matches = epoch_pattern.findall(training_log)

epochs = [int(match[0]) for match in matches]
epoch_losses = [float(match[2]) for match in matches]
latest_time = float(matches[-1][1])

epochs, epoch_losses, latest_time
```

```python
plt.figure(figsize=(10, 6))
plt.plot(epochs, epoch_losses, marker='o')
plt.title("Epoch Loss over Time")
plt.xlabel("Epoch")
plt.ylabel("Epoch Loss")
plt.grid(True, linewidth=0.25)

plt.annotate(f'{epoch_losses[-1]:.4f}',
             (epochs[-1], epoch_losses[-1]),
             textcoords="offset points",
             xytext=(-10,10)
)

plt.tight_layout()
plt.show()
```
